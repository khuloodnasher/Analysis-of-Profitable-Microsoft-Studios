## Introduction 
### Microsoft Industry Project
In an effort by Microsoft to diversify its investments, they decided to invest in the film industry, and since Microsoft is keen on implementing its plans professionally, they have decided to employ a data scientist  whose job is to explore the data of the movie industry, analyze it and set recommendations based on the outcomes of the analysis.

The following Questions were addressed: 

1- How has the movie industry changed in the last decade from 2008-2018?

2- What production budget range is more profitable for movies?

3- Which season is the most profitable?

4- What are the most successful genres?

5- Which genre is the most profitable for each season?

6- Does the runtime have any effect on the popularity of the genre?

### Second: uploading the data

```python
# Uploading the Data Sheet and creating a Data Frame
df_tn = pd.read_csv('tn.movie_budgets.csv.gz')
df_tn.head()
df_tn.info()
```

### Cleaning the Data

```python
# formatting the relase date to a datetime format(YY--mm-dd) 
df_tn['release_date'] = pd.to_datetime(df_tn['release_date'], infer_datetime_format=True)
df_tn.head()
```

```python
# Replacing the '$' sign
df_tn['production_budget'] = df_tn['production_budget'].str.replace('$','')
df_tn['domestic_gross'] = df_tn['domestic_gross'].str.replace('$', '')
df_tn['worldwide_gross'] = df_tn['worldwide_gross'].str.replace('$', '')
df_tn.head()
```

```python
# Delete commas from "production_budget,domestic_gross and worldwide_gross columns"
df_tn['production_budget'] = df_tn['production_budget'].str.replace(',', '')
df_tn['domestic_gross'] = df_tn['domestic_gross'].str.replace(',', '')
df_tn['worldwide_gross'] = df_tn['worldwide_gross'].str.replace(',', '')
df_tn.head()
```

```python
# lowering case the movie column

df_tn['movie'] = df_tn['movie'].str.lower()
df_tn.head()
```

```python
# changing the type of the production budget,domestic gross and worldwide gross to mumeric as integer
df_tn['production_budget'] = df_tn['production_budget'].astype('int64')
df_tn['domestic_gross'] = df_tn['domestic_gross'].astype('int64')
df_tn['worldwide_gross'] = df_tn['worldwide_gross'].astype('int64')

# Dropping unneeded columns
df_budget = df_tn.drop(columns = ['id'])
df_budget.head()
```

```python
# Calcualting the revenue, and profit and adding them to the dataframe
df_tn['Revenue'] = df_tn['domestic_gross'] + df_tn['worldwide_gross']
df_tn['Profit'] = df_tn['Revenue'] - df_tn['production_budget']
df_tn.head()
```

```python
# removing duplicated values from the dataframe
df_tn.drop_duplicates(subset=None, keep='first')
df_tn.duplicated().value_counts()
```

```python
# Uploading the second table 
df_genres = pd.read_csv('imdb.title.basics.csv.gz')
df_genres.info()
```

```python
# removing duplicated values from the dataframe
df_genres.drop_duplicates(subset=None, keep='first')
df_genres.duplicated().value_counts()
```

```python
# Renaming the column original_title
df_genres = df_genres.rename(columns={'original_title' : 'movie'})
df_genres['movie'] = df_genres['movie'].str.lower()
df_genres.head()
```

```python
# uploading third file the tmdb_movies file
df_tmdb_movies = pd.read_csv('tmdb.movies.csv.gz')
df_tmdb_movies.head()
```

```python
# renaming original title to movie
df_tmdb_movies = df_tmdb_movies.rename(columns={'original_title' : 'movie'}) # rename the column
df_tmdb_movies ['movie'] = df_tmdb_movies['movie'].str.lower()
df_tmdb_movies.head()
```

```python
# Removing duplicates
df_tmdb_movies.drop_duplicates(subset=None, keep='first')
df_tmdb_movies.info()
df_tmdb_movies.isna().sum() ## checking Nan values
```

```python
# Dropping unneeeded columns
df_tmdb_movies = df_tmdb_movies.drop(columns = ['Unnamed: 0','id','genre_ids','title','vote_count','original_language','vote_average'])
df_tmdb_movies.head()
```

### Joining Three dataframes

```python
df_merged = df_genres.merge(df_tmdb_movies, on='movie').merge(df_tn, on='movie')
df_merged.drop_duplicates(subset=None, keep='first',inplace=True)
df_merged.duplicated().value_counts()
df_merged.info()
```

```python
# Dropping unneeded columns,''
df_merged = df_merged.drop(columns = ['release_date_x','tconst','primary_title','start_year','id'])
# rename the coloumn
df_merged = df_merged.rename(columns={'release_date_y' : 'release_date'}) 
df_merged.info()
```

```python
## extracting month and year  from release date
df_merged['release_month'] = pd.DatetimeIndex(df_merged['release_date']).month 
df_merged['release_year'] = pd.DatetimeIndex(df_merged['release_date']).year  
```

```python
# Dropping duplicate columns based on movie name and release date
df_merged=df_merged.drop_duplicates(subset=['movie','release_date'], keep='first')
```

```python
## checking duplicated values
df_merged.info()
df_merged.duplicated().value_counts()
```

### Considering the last decade's movies 2008-2018

```python
df_new=df_merged[df_merged['release_year']>=2008]
df_merged=df_new[df_new['release_year']<=2018]
df_merged['release_year'].max() # checking the highest year
df_merged['release_year'].min() # checking the lowest year
df_merged['Profit'].max() # checking the maximum profit
df_merged['Profit'].min() # checking the minimum profit
df_merged = df_merged.sort_values(by='Profit', ascending=False) #Sorting the value by Profit column in descending order
df_merged.info()
df_merged.head()
```

```python
## checking the null values
df_merged.isna().sum()
```

```python
#dropping null values from 'genres column' and ''runtime_minutes''
df_merged = df_merged.dropna(subset=['genres'])
df_merged = df_merged.dropna(subset=['runtime_minutes'])
```

```python
#Extracting the month name from the release date and put it in word format with first three letters
df_merged['Month'] = pd.to_datetime(df_merged['release_date'], format='%m').dt.month_name().str.slice(stop=3)

#Splitting the genres category
df_merged['genres']=df_merged['genres'].str.split(",",3)
df_merged=df_merged.explode('genres')
df_merged.info()

#Sorting the value by Profit column
df_merged = df_merged.sort_values(by='Profit', ascending=False)
```

```python
# set season function according to release_month: Winetr  is the month of December-February,Spring:March-May
# Summer: June-August,Fall:September-November


def seasons(season):
    return season.map(lambda x: "winter/holiday" if x > 11 or x < 3
                      else ("spring" if x >= 3 and x < 6
                            else ("summer" if x > 5 and x < 9
                                  else ("Fall"))))
```

```python
# applying the seasons function and adding season column to assess if release season affects profit
df_merged['release_season'] = seasons(df_merged['release_month'])

df_merged['release_season'].value_counts()
```

```python
# set runtime in hour function according to runtime_minutes column: runtime is defined as less than one hour if < 60 minutes 
# 1-2 hours if 60>runtime>120 and  above 2 hours with any time else

def run_time(runtime):
    return runtime.map(lambda y: "Less than 1 hour" if y < 60
                                         else ("1-2 hours" if y > 60 and y < 120
                                         else ("More than 2 hours")))

df_merged['runtime_hour'] = run_time(df_merged['runtime_minutes']) # applying the runtime function to the dataframe
                                         
df_merged['runtime_hour'].value_counts()  

df_merged.head()
```

### plotting profit by year

```python
# profit versus release year
plt.figure(figsize=(15,8))
plt.style.use('dark_background')
ax1=sns.stripplot(x=df_merged.release_year, y=df_merged.Profit/1000000000, data = df_merged)
ax1.set_title('Profit vs Release Year',fontsize=20,color='white')
ax1.set_xlabel('Release Year',fontsize=15,color='white')
ax1.set_ylabel('Profit (billion USD)',fontsize=15,color='white');
```

### Plotting profit versus  month

```python
### Profit versus release month
df_merged = df_merged.sort_values(by='release_month', ascending=True)
sns.set(rc={'figure.figsize':(15,7)}) #Setting the Chart area size
h=sns.barplot(x=df_merged.Month,y=df_merged.Profit/1000000,color="r")
h.set_xticklabels(h.get_xticklabels(),rotation=90)
h.set_title("Profit by Month", color='black', fontsize=20) 
h.set(xlabel='Month', ylabel='Profit (million USD)')
plt.show();
```

###  Linear regression analysis of Production Budget versus profit

```python
## Linear regression analyses of Production Budget versus profit
plt.figure(figsize=(15,8))
plt.style.use('dark_background')
sns.regplot( x=df_merged.production_budget/1000000, y=df_merged.Profit/1000000000, data=df_merged)
plt.title('Correlation Between Production Budget and Profit', fontsize=20,color='blue')
plt.ylabel('Profit (billion USD)', color='blue')
plt.xlabel('Production Budget (million USD)',color='blue');
```

### Analyzing the profit versus the release season in the last decade

```python
## Grouping by release year, and release season and slicing by the mean of the profit  
## line plot visualizing of the profit versus release season and release year
df_season_profit=(df_merged.groupby(['release_year','release_season'])['Profit'].mean()/1000000).sort_values(ascending=False)
df_season_profit.head()
sns.set_context('talk')
fig = df_season_profit.unstack().plot(kind = 'line')
plt.title('Release Season and Year vs Profit',fontsize=18,color='white')
plt.ylabel('Profit (million USD)',fontsize=15)
fig.legend(loc='lower right', frameon=False);

```

### Investigating the Profit through the genres feature

```python
## Barh plot of profit versus genres
## new dataframe created by grouping by genres and slicing by the  mean of profit in descendant order
df_generes = (df_merged.groupby(['genres'])['Profit'].mean()/1000000).sort_values(ascending=False)
df_generes.head()
plt.style.use('dark_background')
fig = df_generes.plot(kind = 'barh',color='brown')
plt.title('Genre vs Profit',fontsize=18,color='white')
plt.xlabel('Profit (million USD)',fontsize=15)
plt.ylabel('Genre',fontsize=18)
plt.rcParams["figure.figsize"] = [15, 8];
```

```python
##Analyzing the profit to the genres by violin plot and stripplot

df_genre_profit=df_merged[['genres','Profit']]
df_genre_profit.info()
df_genre_profit.sort_values(by=['Profit'],ascending=False)
plt.figure(figsize=(15,8))
sns.set_style('whitegrid')
plt.title("Profit by Genre", fontsize=20)
# Violin plot
ax=sns.violinplot(x=df_genre_profit.genres, y=df_genre_profit.Profit/1000000000, data=df_genre_profit)
# stripplot
ax=sns.stripplot(x=df_genre_profit.genres, y=df_genre_profit.Profit/1000000000, data=df_genre_profit)
plt.xlabel('Genres',fontsize=15,color='blue')
plt.ylabel('Profit (Billion USD)',fontsize=15,color='blue')
plt.xticks(rotation=90);
```
